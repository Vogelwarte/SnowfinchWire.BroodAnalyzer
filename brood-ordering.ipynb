```python
import pandas as pd
import numpy as np
from datetime import timedelta
```

# Austria

```python
austria_df = pd.read_csv('_data/snowfinch-austria.csv')
austria_df
```

```python
austria_df = austria_df[['BroodID', 'Day', 'Month', 'Year', 'Hour', 'Minute', 'Age', 'Nr_nestlings']].dropna()
austria_df
```

```python
austria_df['Age'] = austria_df['Age'].apply(lambda age: age if len(age) < 5 else f'{age[:-4]}.5').astype('float')
austria_df
```

```python
austria_df['Datetime'] = pd.to_datetime(austria_df[['Year', 'Month', 'Day', 'Hour', 'Minute']])
austria_df = austria_df[['BroodID', 'Datetime', 'Age', 'Nr_nestlings']]
austria_df
```

```python
austria_df = austria_df.groupby(['BroodID', 'Datetime']).agg({
	'Age': ['min', 'max'],
	'Nr_nestlings': 'max'
})
austria_df = austria_df.reset_index()
austria_df.columns = ['brood_id', 'datetime', 'age_min', 'age_max', 'brood_size']
austria_df
```

# Switzerland

```python
swiss_df = pd.read_csv('_data/snowfinch-switzerland.csv')
swiss_df
```

```python
def is_brood_id(label: str) -> bool:
	return 'Furka' in label or 'Joch' in label


swiss_df_fmt = pd.DataFrame()

i = 0
while not is_brood_id(swiss_df.loc[i, 'index']):
	i += 1

while i < len(swiss_df):
	brood_id = swiss_df.loc[i, 'index']

	j = 1
	while i + j < len(swiss_df) and not is_brood_id(swiss_df.loc[i + j, 'index']):
		j += 1

	# TODO: NaN handling
	brood_df = swiss_df.iloc[i + 1:i + j].set_index('index').transpose()
	age_cols = [col for col in brood_df.columns if 'Age' in col]
	brood_df = brood_df.replace('?', np.nan).dropna(subset = age_cols, how = 'all').dropna(subset = ['Nr. Nestlings'])

	brood_df['age_min'] = brood_df[age_cols].astype('float').min(axis = 1)
	brood_df['age_max'] = brood_df[age_cols].astype('float').max(axis = 1)

	brood_df = brood_df[['Nr. Nestlings', 'age_min', 'age_max']].reset_index().rename(columns = {
		'index': 'datetime',
		'Nr. Nestlings': 'brood_size'
	})
	brood_df['brood_id'] = brood_id

	swiss_df_fmt = pd.concat([swiss_df_fmt, brood_df])
	i = i + j

swiss_df_fmt = swiss_df_fmt.reset_index().drop(columns = 'index')
swiss_df_fmt['datetime'] = pd.to_datetime(swiss_df_fmt['datetime'], format = '%d/%m/%Y')
swiss_df_fmt['datetime'] += timedelta(hours = 12)
swiss_df_fmt['brood_size'] = swiss_df_fmt['brood_size'].astype('int')
swiss_df_fmt
```
